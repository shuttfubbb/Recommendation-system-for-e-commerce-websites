```python
import pickle
from collections import Counter

with open('yoochoose_64_9556/train_64.txt', 'rb') as f:
    train = pickle.load(f)

with open('yoochoose_64_9556/test_64.txt', 'rb') as f:
    test = pickle.load(f)
```

```python
my_set = set()
cnt_ss_train = 0
cnt_ss_test = 0
total_len = 0

for s in train:
    cnt_ss_train += 1
    total_len += len(s)
    for i in s:
        my_set.add(i)

for s in test:
    cnt_ss_test += 1
    total_len += len(s)
    for i in s:
        my_set.add(i)
```

```python
print(len(my_set))
print(cnt_ss_train)
print(cnt_ss_test)
print(total_len / (len(train) + len(test)))
```

```python
demo = []
for s in test:
    if s[0] <= 20:
        demo.append(s)

print(demo)
```
